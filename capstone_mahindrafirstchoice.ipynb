```
#Customer segmentation based on revenue and comback
repeated_users=pd.pivot_table(jtd_invoice_customer_cleaned,values=['Total_revenue'],index=['User ID','Invoice Date','Order Type','Make'],
               aggfunc={'Total_revenue':'sum','User ID':'count'})
repeated_users=repeated_users.rename(columns={'User ID':'Useridcomeback'})
repeated_users.reset_index(inplace=True)
golden_customers=repeated_users.loc[(repeated_users['Useridcomeback']>25)]
golden_customers_revenue=golden_customers.groupby(['User ID']).agg({'Total_revenue':'sum'})
golden_customers_revenue
loyal_customers=repeated_users.loc[(repeated_users['Useridcomeback']>10)&(repeated_users['Useridcomeback']<=25)]
loyal_customers_revenue=loyal_customers.groupby(['User ID']).agg({'Total_revenue':'sum'})
loyal_customers_revenue
avg_customers=repeated_users.loc[(repeated_users['Useridcomeback']>5)&(repeated_users['Useridcomeback']<=10)]
avg_customers_revenue=avg_customers.groupby(['User ID']).agg({'Total_revenue':'sum'})
avg_customers_revenue
attention_required_customers=repeated_users.loc[(repeated_users['Useridcomeback']<=5)]
attention_required_customers_revenue=attention_required_customers.groupby(['User ID']).agg({'Total_revenue':'sum'})
attention_required_customers_revenue
golden_customers_revenue_total=golden_customers_revenue['Total_revenue'].sum()
loyal_customers_revenue_total=loyal_customers_revenue['Total_revenue'].sum()
avg_customers_revenue_total=avg_customers_revenue['Total_revenue'].sum()
attention_required_customers_revenue_total=attention_required_customers_revenue['Total_revenue'].sum()
print(golden_customers_revenue_total,loyal_customers_revenue_total,avg_customers_revenue_total,attention_required_customers_revenue_total)
```
